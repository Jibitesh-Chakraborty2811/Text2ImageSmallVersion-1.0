```python
X = np.array(X)
Y = np.array(Y)
```

```python
X.shape
```

```python
Y.shape
```

```python
X
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=100,oov_token='<oov>')

tokenizer.fit_on_texts(X)
```

```python
word_index = tokenizer.word_index

word_index
```

```python
X = tokenizer.texts_to_sequences(X)
```

```python
X
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X = pad_sequences(X,maxlen=1,padding='post',truncating='post')
X
```

```python
X.shape
```

```python
X = X.reshape([10,1,1])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Bidirectional,LSTM,Convolution2DTranspose,Reshape,Flatten,Dense

model = Sequential()
#model.add(Embedding(len(tokenizer.word_index),64))
#model.add(Bidirectional(LSTM(512,return_sequences=True)))
model.add(Bidirectional(LSTM(512,return_sequences=True),input_shape=[1,1]))
model.add(Bidirectional(LSTM(512,return_sequences=True)))
model.add(Bidirectional(LSTM(512)))
model.add(Flatten())
model.add(Dense(76800,activation='sigmoid'))
model.add(Reshape([160,160,3]))



model.compile(loss='mse',optimizer='adam',metrics='accuracy')
```

```python
model.summary()
```

```python
model.fit(X,Y,epochs=100)
```

```python
input = 'Deer'
input = [input]
input = tokenizer.texts_to_sequences(input)
input
```
